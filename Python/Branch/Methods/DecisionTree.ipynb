```python
'''
读入NB15_3数据集专用
'''
DATA = '../Datasets/NB15_3.csv'
packets = []
with open(DATA, 'r') as f:
    next(f)
    for line in f:
        key, srcip, dstip, sport, dsport, p_bits, tcpflag, global_time, ttl, label = line.strip().split(',')
        key = int(key)
        sport = int(sport)
        dsport = int(dsport)
        tcpflag = int(tcpflag)
        global_time = int(global_time)
        label = int(label)
        p_bits = int(p_bits)
        ttl = int(ttl)
        packets.append((key, sport, dsport, p_bits, tcpflag, global_time, ttl, label))
```

```python
# 逐包检测
features = []  # 总和, 均值, 最大值, 方差
windowLabel = []  # 标签
for i in range(0, len(packets)):
    key, sport, dsport, p_bits, tcpflag, global_time, ttl, label = packets[i]
    if label == 0 and key % 100 != 0:
        continue

    if label == 0:
        windowLabel.append(0)
        features.append([p_bits])
    elif label in []:
        pass
    else:
        windowLabel.append(1)
        features.append([p_bits])

```

```python
# 朴素贝叶斯
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

X = np.array(features)
y = np.array(windowLabel)

# 定义要过滤的类别
filter_vals = []   # 举例：去掉标签 0 和 3

# 生成保留的索引（即 y 不在 filter_vals 中的行）
mask = ~np.isin(y, filter_vals)

# 应用筛选
X = X[mask]
y = y[mask]

# 划分训练集和测试集（例如 50% 测试）
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.5, random_state=54009
)

# 创建朴素贝叶斯分类器
clf = GaussianNB()

# 拟合训练数据
clf.fit(X_train, y_train)

# 在测试集上预测
y_pred = clf.predict(X_test)

# 输出评估结果
print("准确率: {:.2f}%".format(accuracy_score(y_test, y_pred) * 100))
print("\n分类报告:")
print(classification_report(y_test, y_pred))
```

```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris

def extract_feature_intervals(clf, feature_names=None, max_value=65536):
    tree = clf.tree_
    n_features = tree.n_features

    if feature_names is None:
        feature_names = [f"f{i}" for i in range(n_features)]

    # 收集每个特征的阈值
    thresholds = {i: [] for i in range(n_features)}
    for node_id in range(tree.node_count):
        fid = tree.feature[node_id]
        thr = tree.threshold[node_id]
        if fid >= 0:
            thresholds[fid].append(thr)

    # 生成区间
    intervals = {}
    for fid, thr_list in thresholds.items():
        thr_list = sorted(set(thr_list))
        bounds = [-np.inf] + thr_list + [max_value]
        intervals[fid] = [(bounds[i] + 1 if np.isfinite(bounds[i]) else 0,
                           int(bounds[i+1])) for i in range(len(bounds)-1)]
    return intervals


def encode_value(x, fid, intervals):
    """给定特征值和划分区间，返回其编码索引"""
    for idx, (low, high) in enumerate(intervals[fid]):
        if low <= x <= high:
            return idx
    return None


def extract_lookup_tables(clf, feature_names=None, max_value=65536):
    tree = clf.tree_
    intervals = extract_feature_intervals(clf, feature_names, max_value)

    if feature_names is None:
        feature_names = [f"f{i}" for i in range(tree.n_features)]

    # 遍历叶子节点，构造路径编码
    def recurse(node_id, path_conditions):
        if tree.feature[node_id] == -2:  # 叶子
            class_id = np.argmax(tree.value[node_id])
            # 根据 path_conditions 映射到区间编码
            codes = {}
            for fid, thr, direction in path_conditions:
                # 找出该特征在哪个区间
                if direction == "L":
                    # <= thr
                    code = [idx for idx, (low, high) in enumerate(intervals[fid]) if high <= thr][0]
                else:
                    # > thr
                    code = [idx for idx, (low, high) in enumerate(intervals[fid]) if low > thr][0]
                codes[fid] = code
            return [(codes, class_id)]
        else:
            fid = tree.feature[node_id]
            thr = tree.threshold[node_id]
            left = recurse(tree.children_left[node_id], path_conditions + [(fid, thr, "L")])
            right = recurse(tree.children_right[node_id], path_conditions + [(fid, thr, "R")])
            return left + right

    results = recurse(0, [])

    # 格式化输出
    print("第一阶段（Feature → 区间编码）")
    for fid, ranges in intervals.items():
        fname = feature_names[fid]
        for idx, (low, high) in enumerate(ranges):
            print(f"{low}..{high} : {fname}_ass({idx});")
        print()

    print("第二阶段（编码组合 → 类别）")
    for codes, cls in results:
        key = tuple(codes.get(fid, "_") for fid in range(tree.n_features))
        # print(f"{key} : class {cls}")
        key_str = "(" + ", ".join(str(k) for k in key) + ")"
        print(f"{key_str} : get_result({cls})")

    return intervals, results


# ==== 示例 ====
features_list = ["_sum", "_mean", "_min", "_max", "_var", "_delta", "_ttl_last", "p_bits"]
intervals, lookup = extract_lookup_tables(clf, feature_names=features_list)
```

```python
# 长短期弹跳窗口
from collections import defaultdict

def calculate_window_features(flow_data, window_size, k = 0):
    """计算窗口特征的通用函数"""
    n = min(len(flow_data), window_size)

    _count = 0
    _sum = 0
    _sqsum = 0
    _max = 0
    _min = 0x7fffffff
    _ttl_last = 0
    _ttl_sum = 0
    syn_count = 0
    ack_count = 0
    tcp_rtt = 0
    flag = 0
    jit = 0
    tcpflag_last = 0

    label_count = 0

    for i, (sport, dsport, p_bits, tcpflag, global_time, ttl, label) in enumerate(flow_data[k:n+k]):
        if i == n - 2:  # 倒数第二个包的时间戳
            jit = global_time
        if i == n - 1:  # 最后一个包
            _ttl_last = ttl
            jit = global_time - jit if n >= 2 else 0
            tcpflag_last = tcpflag  # 最后一个包的 flag

        syn_count += 1 if tcpflag ^ 2 else 0
        ack_count += 1 if tcpflag ^ 16 else 0
        tcp_rtt = syn_count + ack_count

        _sum += p_bits
        _max = max(_max, p_bits)
        _min = min(_min, p_bits)
        _sqsum += p_bits ** 2
        _count += 1
        _ttl_sum += ttl

        if label != 0:
            label_count += 1
            flag = label

    # 计算统计值
    _mean = _sum // n if n > 0 else 0
    _sqmean = _sqsum // n if n > 0 else 0
    _var = _sqmean - _mean ** 2
    _delta = _max - _min

    # 获取最后一个包的p_bits
    last_p_bits = flow_data[n-1][2] if n > 0 else 0

    return ([_sum, _mean, _min, _max, _var, tcpflag_last, jit, _count,
             syn_count, ack_count, tcp_rtt, _delta, _ttl_last, last_p_bits, _ttl_sum],
            flag)

# 参数
window_size = 8  # 长期窗口大小
short_window_size = 4  # 短期窗口大小

# 输出 - 长期窗口
features_long = []
windowLabel_long = []

# 输出 - 短期窗口
features_short = []
windowLabel_short = []

features_short_last = []
windowLabel_short_last = []

# 按流存储数据
flow_dict = defaultdict(list)

for pkt in packets:
    key, sport, dsport, p_bits, tcpflag, global_time, ttl, label = pkt
    if label == 0 and key % 100 != 0:
        continue

    # 将包放到该流的列表中
    flow_dict[key].append((sport, dsport, p_bits, tcpflag, global_time, ttl, label))

    # 当前流的所有包
    cur_flow = flow_dict[key]

    # 如果达到一个完整窗口，才计算特征
    if len(cur_flow) == window_size:
        # 计算长期窗口特征（完整的8个包）
        long_features, long_flag = calculate_window_features(cur_flow, window_size)

        # 计算短期窗口特征（前4个包）
        short_features, short_flag = calculate_window_features(cur_flow, short_window_size)

        # 计算短期窗口特征（后4个包）
        short_features_last, short_flag_last = calculate_window_features(cur_flow, short_window_size, 4)

        filter = [9]

        # 存储长期窗口特征和标签
        if label == 0:
            windowLabel_long.append(long_flag)
            features_long.append(long_features)
        elif label in filter:
            pass
        else:
            windowLabel_long.append(long_flag)
            features_long.append(long_features)

        # 存储短期窗口特征和标签
        if label == 0:
            windowLabel_short.append(short_flag)
            features_short.append(short_features)
        elif label in filter:
            pass
        else:
            windowLabel_short.append(short_flag)
            features_short.append(short_features)

        # 存储短期窗口特征和标签
        if label == 0:
            windowLabel_short_last.append(short_flag_last)
            features_short_last.append(short_features_last)
        elif label in filter:
            pass
        else:
            windowLabel_short_last.append(short_flag_last)
            features_short_last.append(short_features_last)

        # 清空当前流，准备下一个窗口
        flow_dict[key] = []
```

```python
# 决策树
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score
import numpy as np

X = np.array(features_long)
y = np.array(windowLabel_long)
# 划分训练集和测试集（例如 80% 训练，20% 测试）
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=54009)
# 创建决策树分类器（可指定 criterion 为 "gini" 或 "entropy"）
clf = DecisionTreeClassifier(criterion='gini', max_depth=None, random_state=54009)
# 拟合训练数据
clf.fit(X_train, y_train)
# 在测试集上预测
y_pred = clf.predict(X_test)



Xs = np.array(features_short)
ys = np.array(windowLabel_short)
# 划分训练集和测试集（例如 80% 训练，20% 测试）
X_train_s, X_test_s, y_train_s, y_test_s = train_test_split(Xs, ys, test_size=0.5, random_state=54009)
# 创建决策树分类器（可指定 criterion 为 "gini" 或 "entropy"）
clf_s = DecisionTreeClassifier(criterion='gini', max_depth=None, random_state=54009)
# 拟合训练数据
clf_s.fit(X_train_s, y_train_s)
# 在测试集上预测
y_pred_s = clf.predict(X_test_s)


Xsl = np.array(features_short_last)
ysl = np.array(windowLabel_short_last)
# 划分训练集和测试集（例如 80% 训练，20% 测试）
X_train_sl, X_test_sl, y_train_sl, y_test_sl = train_test_split(Xsl, ysl, test_size=0.5, random_state=54009)
# 创建决策树分类器（可指定 criterion 为 "gini" 或 "entropy"）
clf_sl = DecisionTreeClassifier(criterion='gini', max_depth=None, random_state=54009)
# 拟合训练数据
clf_sl.fit(X_train_sl, y_train_sl)
# 在测试集上预测
y_pred_sl = clf.predict(X_test_sl)

# 三个预测结果 (假设长度一致)
preds = np.array([y_pred, y_pred_s, y_pred_sl]).T  # (n_samples, 3)

# 判断是否有众数：如果前两个相等，取前两个；否则如果后两个相等取那个；否则全不同取第一个
final_pred = np.where(
    preds[:,0] == preds[:,1], preds[:,0],
    np.where(preds[:,1] == preds[:,2], preds[:,1],
             np.where(preds[:,0] == preds[:,2], preds[:,0], preds[:,0]))
)

# 三个预测结果 (假设长度一致)
predst = np.array([y_test, y_test_s, y_test_sl]).T  # (n_samples, 3)

# 判断是否有众数：如果前两个相等，取前两个；否则如果后两个相等取那个；否则全不同取第一个
final_predt = np.where(
    predst[:,0] == predst[:,1], predst[:,0],
    np.where(predst[:,1] == predst[:,2], predst[:,1],
             np.where(predst[:,0] == predst[:,2], predst[:,0], predst[:,0]))
)


# 输出评估结果
print(classification_report(final_pred, final_predt))
```
